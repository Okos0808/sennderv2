##### Question 1

```python
%%sql
select cast(extract(month from ord_date)as int) as month, office as region, team, 
	SUM( CASE WHEN cancelled = 'No' THEN 1 ELSE 0 END ) as non_cancel_cnt,
	SUM( CASE WHEN cancelled = 'Yes' THEN 1 ELSE 0 END ) as cancel_cnt,
	ROUND(CAST(sum(revenues) as numeric),2) as total_rev,
	ROUND(CAST(sum(cost) as numeric),2) as total_cost,
	ROUND(((CAST(sum(revenues) as numeric)-CAST(sum(cost) as numeric))/CAST(sum(revenues) as numeric))*100,2)  as margin_percent
from opi_dat
group by 1,2,3
order by 1,2,3;
```

##### Question 2

```python
%%sql
select cast(extract(month from ord_date)as int) as month, office as region, team,
	sUM(CASE WHEN gps = 'Yes' THEN 1 ELSE 0 END) as num_of_gpst,
    round((cast(SUM(CASE WHEN gps = 'Yes' THEN 1 ELSE 0 END)as numeric)/cast(SUM( CASE WHEN cancelled = 'No' THEN 1 ELSE 0 END) as numeric))*100,2) as gps_percent,
  	SUM( CASE WHEN ontime = 'Yes' THEN 1 ELSE 0 END ) as number_of_ontime,
  	ROUND(CAST(SUM( CASE WHEN ontime = 'Yes' THEN 1 ELSE 0 END )as numeric)/CAST(SUM( CASE WHEN ontime is not NULL THEN 1 ELSE 0 END ) as numeric)*100,2) as ontimepercent
from opi_dat
group by 1,2,3
order by 1,2,3;
```

##### Question 3

```python
%%sql
select cast(extract(month from ord_date)as int) as month, office as region, team,
	SUM(CASE WHEN market = 'OCTOPUS' and cancelled = 'No' THEN 1 ELSE 0 END) as abs_octopus,
	SUM(CASE WHEN market = 'ORCAS' and cancelled = 'No' THEN 1 ELSE 0 END) as abs_orcas,
	ROUND(CAST(SUM( CASE WHEN market = 'OCTOPUS' and cancelled = 'No' THEN 1 ELSE 0 END )as numeric)/
		  (CAST(SUM( CASE WHEN market = 'OCTOPUS' and cancelled = 'No' THEN 1 ELSE 0 END )as numeric)+
		   CAST(SUM( CASE WHEN market = 'ORCAS' and cancelled = 'No' THEN 1 ELSE 0 END )as numeric))*100,2) AS octo_percent,
	ROUND(CAST(SUM( CASE WHEN market = 'ORCAS' and cancelled = 'No' THEN 1 ELSE 0 END )as numeric)/
		  (CAST(SUM( CASE WHEN market = 'OCTOPUS' and cancelled = 'No' THEN 1 ELSE 0 END )as numeric)+
		   CAST(SUM( CASE WHEN market = 'ORCAS' and cancelled = 'No' THEN 1 ELSE 0 END )as numeric))*100,2) AS orca_percent
from opi_dat
group by 1,2,3
order by 1,2,3;
```
